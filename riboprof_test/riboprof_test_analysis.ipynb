```python
import pandas as pd 
import xpresstools as xp
import scipy.stats as stats
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

### Get data and clean up formatting

```python
data = pd.read_csv('riboprof_test_counts_table.tsv',sep='\t', index_col=0)
data.columns = data.columns.str.replace('trimmed_', '')
data.columns = data.columns.str.replace('_1_final', '')

data.head()
```

### Threshold

```python
data_t = data.loc[data.max(axis=1) > 25]

print(str(data.shape[0]) + ' genes before thresholding')
print(str(data_t.shape[0]) + ' genes before thresholding')
```

### TPM Normalize

```python
data_n = xp.tpm(data_t, './Mus_musculus.GRCm38.95.gtf', gene_name_prefix='gene_id \"', gene_name_location=0, sep='\t')

data_a = xp.tpm(data, './Mus_musculus.GRCm38.95.gtf', gene_name_prefix='gene_id \"', gene_name_location=0, sep='\t')


data_n.head()
```

### Get metadata

```python
meta = pd.read_csv('SraRunTable.txt',sep='\t')
meta = meta[['Run','type']]
meta.columns = [0,1]

meta.head()
```

### Check replicate and RPF to RNA correlations (Fig 1E)

```python
data_c = data_a.copy()

gene_a = data_c[['SRR7777908','SRR7777909']].sum(axis=1).values.tolist()
gene_a = [x + 1 for x in gene_a]
gene_a = np.array(gene_a).astype(np.float)
gene_a = np.ndarray.tolist(gene_a)

gene_b = data_c[['SRR7777895','SRR7777896','SRR7777897']].sum(axis=1).values.tolist()
gene_b = [x + 1 for x in gene_b]
gene_b = np.array(gene_b).astype(np.float)
gene_b = np.ndarray.tolist(gene_b)

gene_a = np.log2(gene_a)
gene_b = np.log2(gene_b)

slope, intercept, r_value, p_value, std_err = stats.linregress(gene_a, gene_b)
x = np.linspace(np.amin(gene_a), np.amax(gene_b), 100)
y = (slope * x) + intercept

fig, ax = plt.subplots(1,1, figsize=(4,4))

ax.scatter(gene_a, gene_b, s=1,c='black')
ax.set_title('R = ' + str(r_value) + '\np-value ' + str(p_value))
ax.plot(x, y, '-r')

plt.show()
```

### Run analyses

```python
combined = data_n.copy()

combined['ribo_seq_wt_plus'] = combined[['SRR7777908','SRR7777909']].sum(axis=1)
combined['ribo_seq_wt_minus'] = combined[['SRR7777910','SRR7777911']].sum(axis=1)
combined['ribo_seq_ko_plus'] = combined[['SRR7777912','SRR7777913']].sum(axis=1)
combined['ribo_seq_ko_minus'] = combined[['SRR7777914','SRR7777915']].sum(axis=1)

combined['rna_seq_wt_plus'] = combined[['SRR7777895','SRR7777896','SRR7777897']].sum(axis=1)
combined['rna_seq_wt_minus'] = combined[['SRR7777898','SRR7777899','SRR7777900']].sum(axis=1)
combined['rna_seq_ko_plus'] = combined[['SRR7777901','SRR7777902','SRR7777903']].sum(axis=1)
combined['rna_seq_ko_minus'] = combined[['SRR7777904','SRR7777905','SRR7777906','SRR7777907']].sum(axis=1)

cols = [c for c in combined.columns if c.upper()[:3] != 'SRR']

combined = combined[cols]

combined.head()
```

```python
combined_te = combined.copy()

combined_te['wt_plus'] = combined_te['ribo_seq_wt_plus'] / combined_te['rna_seq_wt_plus']
combined_te['wt_minus'] = combined_te['ribo_seq_wt_minus'] / combined_te['rna_seq_wt_minus']
combined_te['ko_plus'] = combined_te['ribo_seq_ko_plus'] / combined_te['rna_seq_ko_plus']
combined_te['ko_minus'] = combined_te['ribo_seq_ko_minus'] / combined_te['rna_seq_ko_minus']

cols = [c for c in combined_te.columns if '_seq_' not in c]

combined_te = combined_te[cols]

combined_te.head()
```

```python
combined_logte = combined.copy()

combined_logte = combined_logte + 1

combined_logte['wt_plus'] = combined_logte['ribo_seq_wt_plus'] / combined_logte['rna_seq_wt_plus']
combined_logte['wt_minus'] = combined_logte['ribo_seq_wt_minus'] / combined_logte['rna_seq_wt_minus']
combined_logte['ko_plus'] = combined_logte['ribo_seq_ko_plus'] / combined_logte['rna_seq_ko_plus']
combined_logte['ko_minus'] = combined_logte['ribo_seq_ko_minus'] / combined_logte['rna_seq_ko_minus']

combined_logte = np.log2((combined_logte))

cols = [c for c in combined_logte.columns if '_seq_' not in c]

combined_logte = combined_logte[cols]

combined_logte.head()
```
